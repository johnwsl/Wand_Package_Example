# Função - Balanceamento/Data Augmentation (tipo random) de Dataset - Flipping imagens (vertical e horizontal), shearing e Rotação.

## Será usado a bliblioteca Wand. Esta por sua vez, necessita do software ImageMagick instalado. Info: https://docs.wand-py.org/en/0.6.7/

```python
from wand.image import Image
import os
import random
```

```python
# A função balance_aug salva na pasta de saída uma quantidade de imagens adicionaias necessárias para o balanceamento 
# imgfileIn -> pasta com dados para balanceamento (classe minoritária)
# imgfilenumber -> número de imagens totais (imagens existentes mais a quantidade que deseja aumentar)
# imgfileOut -> pasta de saída, com imagens adicionais para balanceamento
# formato -> formato das imagens de entrada da pasta imgfileIn
# Observação: o número imgfilenumber deve ser sempre maior que a quantidade de elementos da pasta imgfileIn que queremos balancear
def balance_aug(imgfileIn,imgfilenumber,imgfileOut,formato='.png'):
            
    walkdir = os.walk(imgfileIn, topdown=False)
    for root, dirs, files in walkdir:
        if len(files) != 0:
            print("Número de Imagens", "no diretório", root,":",len(files), "\n")
            
            x = imgfilenumber - len(files)
                
            filesrandom = files
            while filesrandom == files:
                filesrandom = random.sample(files,k=len(files))
                
            xfinal= 0  
            if x>len(filesrandom):
                xfinal = x - len(filesrandom)
                
            filesrandomFinal = filesrandom
            while filesrandomFinal == filesrandom:
                filesrandomFinal = random.sample(filesrandom,k=xfinal)
            
            print("Número de imagens adicioais após balanceamento (imagens adicionadas na pasta de saída): ",
                  (len(filesrandom)+xfinal),"\n")
            print("Número total de imagens após balanceamento: ",(len(filesrandom)+xfinal+len(files)),"\n")
            
            if len(filesrandom)%2==0:
                y = int(len(filesrandom)/2)
            else:
                y = int((len(filesrandom)+1)/2)
                         
            count = 0 
            for name in (filesrandom[:y]):
                count = count + 1
                with Image(filename = imgfileIn + name) as img:
                    with img.clone() as flip:
                        flip.flip()
                        flip.rotate(random.choices([0,40,90,120,160],weights = [1,1,1,1,1], k = 1)[0], 'black')
                        flip.shear('black', random.choices([0,10,20],weights = [1,1,1], k = 1)[0], 
                                   random.choices([0,10,20],weights = [1,1,1], k = 1)[0])
                        flip.save(filename = imgfileOut+"img"+str(count)+formato)
                        print("Img -",count,"-",name,"- Fase1")
            for name in (filesrandom[y:len(filesrandom)]):
                count = count + 1
                with Image(filename = imgfileIn + name) as img:
                    with img.clone() as flop:
                        flop.flop()
                        flop.rotate(random.choices([0,40,90,120,160],weights = [1,1,1,1,1], k = 1)[0], 'black')
                        flop.shear('black', random.choices([0,10,20],weights = [1,1,1], k = 1)[0], 
                                   random.choices([0,10,20],weights = [1,1,1], k = 1)[0])
                        flop.save(filename = imgfileOut+"img"+str(count)+formato)
                        print("Img -",count,"-",name,"- Fase2")
                        
            if len(filesrandomFinal)%2==0:
                y = int(len(filesrandomFinal)/2)
            else:
                y = int((len(filesrandomFinal)+1)/2)
                        
                        
            if (xfinal!=0):
                for name in (filesrandomFinal[:y]):
                    count = count + 1
                    with Image(filename = imgfileIn + name) as img:
                        with img.clone() as flip:
                            flip.flip()
                            flip.rotate(random.choices([0,40,90,120,160],weights = [1,1,1,1,1], k = 1)[0], 'black')
                            flip.shear('black', random.choices([0,10,20],weights = [1,1,1], k = 1)[0], 
                                       random.choices([0,10,20],weights = [1,1,1], k = 1)[0])
                            flip.save(filename = imgfileOut+"img"+str(count)+formato)
                            print("Img -",count,"-",name,"- Fase1")
                for name in (filesrandomFinal[y:]):
                    count = count + 1
                    with Image(filename = imgfileIn + name) as img:
                        with img.clone() as flop:
                            flop.flop()
                            flop.rotate(random.choices([0,40,90,120,160],weights = [1,1,1,1,1], k = 1)[0], 'black')
                            flop.shear('black', random.choices([0,10,20],weights = [1,1,1], k = 1)[0], 
                                       random.choices([0,10,20],weights = [1,1,1], k = 1)[0])
                            flop.save(filename = imgfileOut+"img"+str(count)+formato)
                            print("Img -",count,"-",name,"- Fase2")
```

```python
balance_aug('BreakHis400x/train_set/benign/', 1037, 'BreakHis400x/train_set/train_benign_aug/')
```
