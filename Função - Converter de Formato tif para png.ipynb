# Função - Converter de Formato .tif para .png

## Será usado a bliblioteca Wand. Esta por sua vez, necessita do software ImageMagick instalado. Info: https://docs.wand-py.org/en/0.6.7/

```python
from wand.image import Image
import os
```

```python
def tif_png(imgfileIn,imgfileOut):
    walkdir = os.walk(imgfileIn, topdown=False)
    for root, dirs, files in walkdir:
        if len(files) != 0:
            print("Número de Imagens", "no diretório", root,":",len(files), "\n")
            count = 0
            for name in files:
                count = count + 1
                with Image(filename = imgfileIn + name) as img:
                    print("Img",count,"- Formato atual: ",img.format, "\n")
                    img.format = "png"
                    img.save(filename = imgfileOut+name.replace(".tif", ".png"))
                    print("Img",count,"- Formato posterior: ",img.format, "\n")
                
```

```python
tif_png('BreCaHAD_dataset/images/', 'BreCaHAD_dataset/BreCaHAD_test_png/')
```
